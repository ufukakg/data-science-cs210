# <h3> Data manipulation (Object to numericals) 

```python
def handle_non_numerical_data(df):
    columns = df.columns.values
    
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            print(column)
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x=0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    #print(unique + " : " + str(x))
                    text_digit_vals[unique] = x
                    x+=1        
            df[column] = list(map(convert_to_int, df[column])) 
    
    return df
```

```python
for i in range(13,32):
    drug_data[drug_data.columns[i]] = drug_data[drug_data.columns[i]].str.replace('CL','').astype(int)
```

```python
for i in range(6,13):
    drug_data[drug_data.columns[i]] = drug_data[drug_data.columns[i]].str.replace(',','.').astype(np.float64)
```

```python
sample_columns = drug_data[['Age','Nicotine']]
```

```python
age_nscore = drug_data[['Gender','Nscore']]
nscore = drug_data[['Nscore']]
nscore_age2 = age_nscore[drug_data["Gender"]==1]
nscore2 = nscore_age2[['Nscore']]
print(nscore.mean())
print(nscore2.mean())
```

```python
print(len(nscore2))
```

```python
stats.ttest_1samp(a= nscore2,               # Sample data
                 popmean= nscore.mean())  # Pop mean
```

```python
stats.t.ppf(q=0.025,  # Quantile to check
            df=941)  # Degrees of freedom
```

```python
stats.t.ppf(q=0.975,  # Quantile to check
            df=941)  # Degrees of freedom
```

```python
stats.t.cdf(x= -1.8010,      # T-test statistic
               df= 941) * 2   # Mupltiply by two for two tailed test*
```

```python
sigma = nscore2.std()/math.sqrt(942)  # Sample stdev/sample size

stats.t.interval(0.95,                        # Confidence level
                 df = 941,                     # Degrees of freedom
                 loc = nscore2.mean(),        # Sample mean
                 scale= sigma)                # Standard dev estimate              # Standard dev estimate
```

```python
stats.t.interval(0.99,                        # Confidence level
                 df = 941,                     # Degrees of freedom
                 loc = nscore2.mean(),        # Sample mean
                 scale= sigma)                # Standard dev estimate    
```

```python
nico_freq = pd.value_counts(sample_columns['Nicotine'])
nico_freq/len(sample_columns)
```

```python
nicotine_table = pd.crosstab(index=sample_columns['Nicotine'], columns=["count"])
nico_ratio = nicotine_table/len(drug_data)
nico_ratio
```

```python
age_0 = sample_columns[drug_data["Age"]==2]
nico_0 = pd.crosstab(index=age_0['Nicotine'], columns=["count"])
nico_0_ratio = nico_0/len(age_0)
nico_0_ratio
```

```python
observed = nico_0
general_ratios = nico_ratio
expected = general_ratios * len(age_0)
chi_squared_stat = (((observed-expected)**2)/expected).sum()
print(chi_squared_stat)
```

```python
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 6)   # Df = number of variable categories - 1

print("Critical value")
print(crit)

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Find the p-value
                             df=6)
print("P value")
print(p_value)
```

```python
stats.chisquare(f_obs= observed,   # Array of observed counts
                f_exp= expected)   # Array of expected counts
```
