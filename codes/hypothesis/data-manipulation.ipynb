# <h3> Data manipulation (Object to numericals) 

```python
def handle_non_numerical_data(df):
    columns = df.columns.values
    
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            print(column)
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x=0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    print(unique + " : " + str(x))
                    text_digit_vals[unique] = x
                    x+=1        
            df[column] = list(map(convert_to_int, df[column])) 
    
    return df

```

```python
for i in range(13,32):
    drug_data[drug_data.columns[i]] = drug_data[drug_data.columns[i]].str.replace('CL','').astype(int)
```

```python
for i in range(2,13):
    drug_data[drug_data.columns[i]] = drug_data[drug_data.columns[i]].str.replace(',','.').astype(np.float64)
```

```python
def change_age(cls):
    if cls == -0.95197:
        return '18-24'
    elif cls == -0.07854:
        return '25-34'
    elif cls == 0.49788:
        return '35-44'
    elif cls == 1.09449:
        return '45-54'
    elif cls == 1.82213:
        return '55-64'
    elif cls == 2.59171:
        return '65+'
    
def change_gender(cls):
    if cls == 0.48246:
        return "male"
    elif cls == -0.48246:
        return "female"

def change_country(cls):
    if cls == -0.09765:
        return "Australia"
    elif cls == 0.24923:
        return "Canada"
    elif cls == -0.46841:
        return "New Zealand"
    elif cls == -0.28519:
        return "Other"
    elif cls == 0.21128:
        return "Rep.of Ireland"
    elif cls == 0.96082:
        return "UK"
    elif cls == -0.57009:
        return "USA"
```

```python
# drug_data['Age'] = drug_data['Age'].str.replace(',','.').astype(np.float64)
drug_data['Age'] = drug_data['Age'].apply(change_age)
```

```python
# drug_data['Gender'] = drug_data['Gender'].str.replace(',','.').astype(np.float64)
drug_data['Gender'] = drug_data['Gender'].apply(change_gender)
```

```python
drug_data['Country'] = drug_data['Country'].apply(change_country)
```

```python
samples = drug_data[['Chocolate','Ecstasy']]
```

```python
ecs_freq = pd.value_counts(samples['Ecstasy'])
ecs_freq/len(samples)
```

```python
ecs_table = pd.crosstab(index=samples['Ecstasy'], columns=["count"])
ecs_ratio = ecs_table/len(drug_data)
ecs_ratio
```

```python
choco_0 = samples[drug_data["Chocolate"]==6]
ecs_0 = pd.crosstab(index=choco_0['Ecstasy'], columns=["count"])
ecs_0_ratio = ecs_0/len(choco_0)
ecs_0_ratio
```

```python
observed = ecs_0
general_ratios = ecs_ratio
expected = general_ratios * len(choco_0)
chi_squared_stat = (((observed-expected)**2)/expected).sum()
print(chi_squared_stat)
```

```python
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 6)   # Df = number of variable categories - 1

print("Critical value")
print(crit)

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Find the p-value
                             df=6)
print("P value")
print(p_value)
```
