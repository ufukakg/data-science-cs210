```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.drop(['PassengerId', 'Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
df
```

```python
df['companions'] = df['SibSp'] + df['Parch']
df
```

```python
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df
```

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,3))
sns.kdeplot(df["Age"][df.Survived == 1], color="darkturquoise", shade=True)
sns.kdeplot(df["Age"][df.Survived == 0], color="lightcoral", shade=True)
plt.legend(['Survived', 'Died'])
plt.title('Density Plot of Age for Surviving Population and Deceased Population')
plt.show()
```

```python
plt.figure(figsize=(15,3))
sns.kdeplot(df["Fare"][df.Survived == 1], color="darkturquoise", shade=True)
sns.kdeplot(df["Fare"][df.Survived == 0], color="lightcoral", shade=True)
plt.legend(['Survived', 'Died'])
plt.title('Density Plot of Fare for Surviving Population and Deceased Population')
# limit x axis to zoom on most information. there are a few outliers in fare. 
plt.xlim(-20,200)
plt.show()
```

```python
def change_class(cls):
    if cls == 1:
        return 'first'
    elif cls == 2:
        return 'second'
    else:
        return 'third'
```

```python
df['Pclass'] = df['Pclass'].apply(change_class)
#pclass_map = {1: 'Upper Class', 2: 'Middle Class', 3: 'Lower Class'}
#df['Pclass'] = df['Pclass'].map(pclass_map)
```

```python
df
```

```python
compare_attr = 'Sex'
df_pclass = df[[compare_attr, 'Survived']].groupby(compare_attr).mean()
df_pclass.plot.bar()
```

```python
pd.crosstab(index=[df['Survived']], columns=df['Sex'])
```

```python
result = pd.crosstab(index=[df['Sex'], df['Pclass'], df['Age']],
            columns=df['Survived'],
           normalize='index')
result
```

```python
table = pd.crosstab(index=[df['Survived']], columns=df['Sex'], normalize='columns')
table
```

```python
test_result = stats.chi2_contingency(table)
```

```python
test_result
```

```python
def binarize_gender(gender):
    if gender == 'female':
        return 1
    else:
        return 0
```

```python
df['Sex'] = df['Sex'].apply(binarize_gender)
```

```python
corr = df.corr()
```
